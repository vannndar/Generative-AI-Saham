```python
import yfinance as yf

# Define the ticker symbol
ticker_symbol = "BBCA.JK"

# Create a Ticker object
ticker = yf.Ticker(ticker_symbol)

# Fetch historical market data
historical_data = ticker.history(period="5y")  # data for the last year
print("Historical Data:")
print(historical_data)

# Fetch basic financials
financials = ticker.financials
print("\nFinancials:")
print(financials)

# Fetch stock actions like dividends and splits
actions = ticker.actions
print("\nStock Actions:")
print(actions)
```

```python
# tulis historical data ke file dan overwrite
historical_data.to_csv('historical_data.csv', index=False)
```

```python
# Pastikan kolom "Date" diubah menjadi datetime tanpa timezone
historical_data.index = pd.to_datetime(historical_data.index).tz_localize(None)

# tambahkan column date_str
historical_data['date_str'] = ''

# Loop untuk mencetak tanggal dalam format "YYYY-MM-DD" dan nilai "Close"
for i in range(len(historical_data)):
    date_str = historical_data.index[i].strftime('%Y-%m-%d')
    close_value = historical_data['Close'][i]
    # tambah kolom baru date_str as datetime dengan data date_str
    historical_data['date_str'][i] = pd.to_datetime(date_str)
    print(date_str, close_value)
```

```python
# hapus 4 baris pertama
weekly_data = historical_data[4:]

weekly_data.head(30)
```

```python

startDate = weekly_data['date_str'].iloc[0]
endDate = startDate + pd.DateOffset(days=7)
print(weekly_data['date_str'].iloc[0].strftime('%Y-%m-%d') == startDate.strftime('%Y-%m-%d'))
print(startDate,endDate)
```

```python
from math import floor


max = len(weekly_data)
last = weekly_data['date_str'].iloc[0]
realEndDate = endDate
output = weekly_data[:]
for i in range(1,max):
    if (weekly_data['date_str'].iloc[i].strftime('%Y-%m-%d')  == realEndDate.strftime('%Y-%m-%d')):
        endDate = realEndDate
    elif (last.strftime('%Y-%m-%d') < endDate.strftime('%Y-%m-%d') and weekly_data['date_str'].iloc[i].strftime('%Y-%m-%d')  > endDate.strftime('%Y-%m-%d')):
        endDate = weekly_data['date_str'].iloc[i]
    if(weekly_data['date_str'].iloc[i].strftime('%Y-%m-%d')  == endDate.strftime('%Y-%m-%d')  ):
        startDate = endDate
        endDate = startDate + pd.DateOffset(weeks=1)
        realEndDate = realEndDate + pd.DateOffset(weeks=1)
        print(weekly_data.index[i].strftime('%Y-%m-%d'), startDate.strftime('%Y-%m-%d'), endDate.strftime('%Y-%m-%d'), realEndDate.strftime('%Y-%m-%d'), weekly_data.index[i] - realEndDate, (weekly_data.index[i] - realEndDate) == pd.Timedelta(0))
        if((endDate - realEndDate)>= pd.Timedelta(0)):
            selisih = endDate - realEndDate
            if(floor(selisih.days/7) > 0):
                realEndDate = realEndDate + pd.DateOffset(weeks=floor(selisih.days/7))
    else:
        output = output.drop(weekly_data.index[i])
        last = weekly_data['date_str'].iloc[i]
        # print(weekly_data.index[i], last.strftime('%Y-%m-%d'), endDate.strftime('%Y-%m-%d'))
```

```python
fig = px.line(historical_data, x=historical_data.index, y="Close", title='Historical Stock Price')
fig.show()
```

```python
fig = px.line(output, x=output.index, y="Close", title='Weekly Stock Price')
fig.show()
```
